```python
def computeJ(v, R0, s0, t0):
    if v.shape[0] != 3:
        raise ValueError("v should have shape 3xN")
    n_points = v.shape[1]
    
    #unreferenced parameter t0
    
    m1 = np.array([[1],[1],[1]])
    
    # i consider all of them matrices
    J = np.zeros((3*n_points, 9 + 3 + 1)) #9 from rotation, 3 from translation, 1 from scale
    for i in range(n_points):
        v_i = v[:,[i]]
        dR = s0 * (m1.dot(v_i.T))          # shape 3x3 
        ds = np.sum(R.dot(v_i))             # shape 1x1  R.dot(v_i) is shape 3x1
        ds = np.ones((3,1)) * ds
        J[3*i+0, 0:3]  = dR.flatten()[0:3]
        J[3*i+1, 3:6]  = dR.flatten()[3:6]
        J[3*i+2, 6:9]  = dR.flatten()[6:9]
        
        J[3*i+0, 9:12] = np.array([1, 0 ,0])
        J[3*i+1, 9:12] = np.array([0, 1 ,0])
        J[3*i+2, 9:12] = np.array([0, 0 ,1])
        
        J[3*i:3*i+3, 12:13] = np.ones((3,1)) * ds
    return J

v = measured_points
R0 = np.eye(3)
s0 = np.array([[1]])
t0 = np.zeros((3,1))
J = computeJ(v, R0, s0, t0)
print J
    
```
