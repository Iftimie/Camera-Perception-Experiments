```python
def rotationToAngularVel(R0, R1, t0, t1):
    #https://math.stackexchange.com/questions/668866/how-do-you-find-angular-velocity-given-a-pair-of-3x3-rotation-matrices
    A = R1.dot(R0.T)
    theta = np.arccos((np.trace(A)-1.0)/2.0)
    
    W_skew =  1.0/(2.0 * (t1-t0)) * (theta/sin(theta)) * (A - A.T)
    return W_skew
    
R0 = np.eye(3)
R1 = RotX(-0.2).dot(RotY(-0.1)).dot(RotZ(-0.2))

w = rotationToAngularVel(R0, R1, 0, 1)
print "rotationToAngularVel\n", w
```
